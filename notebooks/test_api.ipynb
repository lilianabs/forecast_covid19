```python
import requests
import pandas as pd
```

```python
url_api = f"https://api.covidtracking.com/v1/us/daily.json"
```

```python
response = requests.get(url_api)
```

```python
# Validate response code
assert(response.status_code == 200)
```

```python
data = response.json()
```

```python
df = pd.DataFrame.from_dict(data)
```
